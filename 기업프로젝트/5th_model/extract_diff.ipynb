```python
import pandas as pd
```

```python
ELEC_DATA_PATH = '../Data/electric_df_28_days.csv'
WATER_DATA_PATH = '../Data/water_df_28_days.csv'
```

```python
electric_df = pd.read_csv(ELEC_DATA_PATH, index_col = 0)
water_df = pd.read_csv(WATER_DATA_PATH, index_col = 0)

electric_df.head()
```

```python
def extract_diff(df):

    diff_df = pd.DataFrame(columns = range(27))

    for i in range(len(df)):
        L = []

        for j in range(27):
            L.append(df.iloc[i][j] - df.iloc[i][j + 1])

        diff_df.loc[i] = L
    
    return diff_df  
```

```python
water_diff_df = extract_diff(water_df)
electric_diff_df = extract_diff(electric_df)
```

```python
electric_diff_df.to_csv('../Data/electric_diff_df.csv', index = 0)
water_diff_df.to_csv('../Data/water_diff_df.csv', index = 0)
```
