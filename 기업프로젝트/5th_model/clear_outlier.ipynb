```python
import pandas as pd
```

```python
ELEC_DATA_PATH = '../Data/electric_df_28_days.csv'
WATER_DATA_PATH = '../Data/water_df_28_days.csv'
```

```python
electric_df = pd.read_csv(ELEC_DATA_PATH)
water_df = pd.read_csv(WATER_DATA_PATH)

electric_df.head()
```

```python
def clear_outlier(df):

    idx = set()

    for i in range(len(df)):
        for j in range(len(df.iloc[0]) - 2):
            if ((df.iloc[i][j] * 0.9) <= df.iloc[i][j + 1] <= (df.iloc[i][j] * 1.1)) and ((df.iloc[i][j + 1] * 0.9) <= df.iloc[i][j + 2] <= (df.iloc[i][j + 1] * 1.1)):
                idx.add(i)

    return idx
```

```python
idx = clear_outlier(water_df)
```

```python
# electric_df.drop(idx, axis = 0, inplace = True)
water_df.drop(idx, axis = 0, inplace = True)
```
