```python
import pandas as pd
```

```python
ELEC_DATA_PATH = '../Data/electric_df.csv'
WATER_DATA_PATH = '../Data/water_df.csv'
```

```python
electric_df = pd.read_csv(ELEC_DATA_PATH, index_col = 0)
water_df = pd.read_csv(WATER_DATA_PATH, index_col = 0)

electric_df.head()
```

```python
idx = set()

for i in range(len(electric_df)):
    for j in range(len(electric_df.iloc[0]) - 2):
        if (electric_df.iloc[i][j] == electric_df.iloc[i][j + 1]) and (electric_df.iloc[i][j] == electric_df.iloc[i][j + 2]):
            idx.add(i)
```

```python
idx
```

```python
electric_df.drop(idx, axis = 0, inplace = True)
water_df.drop(idx, axis = 0, inplace = True)
```
