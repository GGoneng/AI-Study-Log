- [3] 데이터 전처리 
    * 기본 확인 파악된 내용들 처리
    * 불필요 데이터 삭제 : 0번 행 삭제
    * 결측치 데이터 처리 : '전출지별' 컬럼 처리
    * 이상치 데이터 처리 : '-'에 대한 처리
    * 데이터 자료형 일치 : 형 변환

```python
# [3-1] 불필요 데이터 처리
copyDF = dataDF.drop(index = 0)
print(f" {copyDF.head()}")
```

```python
# [3-2] 결측치 데이터 처리 => isna()/isnull()
copyDF.isna().sum()
```

```python
# - '1984', '1985', '1986' 컬럼에서 결측치 데이터만 추출
print(copyDF.columns)
a = copyDF[[1984, 1985, 1986]].isna()
a[a.sum(axis = 1) > 0]
```

```python
# - '전출지별'
copyDF["전출지별"].ffill(inplace = True)
copyDF[[1984, 1985, 1986]] = copyDF[[1984, 1985, 1986]].fillna(0)
copyDF.loc[35]
```

```python
df2 = pd.DataFrame([[1, 1, pd.NA], [pd.NA, pd.NA, 2], [pd.NA, 3, 3]])
df2
```

```python
# 0번 컬럼을 이전 행 값으로 채우기
df2[0]
```

```python
a = df2[0].ffill(inplace = True)
print(a, df2, sep = "\n")
```

```python
df2[[1, 2]].ffill(inplace = True)
```

```python
# [3-3] 이상치 데이터 처리
copyDF.replace("-", 0, inplace = True)
```

```python
# [3-4] 자료형 변환 
d_ = dict(zip(copyDF.columns[2:].to_list(), ['int'] * (copyDF.shape[1]-2)))

copyDF = copyDF.astype(d_)
```

```python
print(f"copyDF.dtypes \n {copyDF.dtypes}")
```

#### [4] 데이터 가공
    * 대구에서 전지역 전입 인구수 분석
    * 전출지별 - '대구광역시'
    * 전입지별 - '대구광역시', '전국' 제외

```python
# - 대구광역시 데이터만 추출
mask = (copyDF['전출지별'] == "대구광역시") 
cityDF = copyDF[mask]
cityDF
```

```python
# - 전출지별 컬럼 삭제
cityDF = cityDF.drop("전출지별", axis = 1)

# - 전입지별 커럼 행 인덱스 설정
cityDF.set_index("전입지별", inplace = True)
```

```python
# - 대구광역시, 전국 행 삭제
cityDF.drop(["대구광역시", "전국"], inplace = True)
```

```python
# - 가공데이터 확인
print(cityDF.dtypes, cityDF.columns, cityDF.index, sep = "\n\n")
```

 [5] 히스토그램으로 특정 컬럼 데이터 분포 시각화


```python
# - 2000년도 데이터의 분포
cityDF.columns
```

```python
# 값 기준 정렬 ==> [기본] 오름차순
people = cityDF[2000].sort_values(ascending = False)
people
```

```python
# 2000년도 대구 지역에서 타 지역으로 이동 인구 수 분포

plt.hist(x = cityDF[2000], bins = 15, histtype = "stepfilled")
plt.hist(x = cityDF[2017], bins = 15, histtype = "step")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 5))

ax1, ax2 = fig.subplots(1, 2, sharey = True)

ax1.hist(x = cityDF[2000], bins = 15, histtype = "stepfilled")
ax2.hist(x = cityDF[2017], bins = 15, histtype = "step")

plt.show()
```

```python
n ,bin, obj = plt.hist(cityDF[2000], bins = 15, histtype = "stepfilled")
print(f'n => {n}')
print(f'bin => {bin.tolist()}')
labels = list(map(int, bin.tolist()))
plt.xticks(bin[::2], labels[::2], rotation = 45)

plt.show()
```
