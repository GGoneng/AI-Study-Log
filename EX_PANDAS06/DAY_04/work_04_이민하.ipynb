```python
"""
# 1. 등급 나누기 
# 2. 가격에 동승자와 본인의 인원 수를 나눠 1인당 가격 전처리 (총 인원, 1인당 좌석 가격 열 생성)
# 3. 도시별 인원수 체크
# 4. 도시에 어느 비율로 고위직이 많은지 체크
# 5. 도시마다 다른 티켓 비용을 비교하여 어느 도시가 가장 물가가 높은지 체크
# 6. 점수를 계산하여 순위 데이터프레임 생성
"""
import pandas as pd
```

```python
file_name = "titanic.csv"
```

```python
DF = pd.read_csv(file_name)
```

```python
DF
```

```python
# 결측치 개수 측정
print(DF.info())
DF.isna().sum()
```

```python
# 결측치 제거
DF2 = DF.dropna(axis = 0, subset = ["embark_town"]).copy()
DF2.isna().sum()
```

```python
# 도시 별 탑승자 수
S_population = DF2["embark_town"].value_counts()[0]
C_population = DF2["embark_town"].value_counts()[1]
Q_population = DF2["embark_town"].value_counts()[2]
```

```python
# 직위 열 구하기
grade = pd.cut(DF2["pclass"], 
               bins = 3, 
               labels = ["상", "중", "하"])
DF2["grade"] = grade
DF2
```

```python
# 입장 인원 수 열 추가하기
total_passengers = DF2["sibsp"] + DF2["parch"] + 1
DF2["total_passengers"] = total_passengers
DF2
```

```python
# 1인당 요금 열 추가하기
per_fare = DF2["fare"] / DF2["total_passengers"]
DF2["per_fare"] = per_fare
DF2
```

```python
# 고위층 비율 구하기

S_high_percentage = ((DF2["grade"] == "상") & (DF2["embarked"] == "S")).sum() / (DF2["embarked"] == "S").sum() * 100
C_high_percentage = ((DF2["grade"] == "상") & (DF2["embarked"] == "C")).sum() / (DF2["embarked"] == "C").sum() * 100
Q_high_percentage = ((DF2["grade"] == "상") & (DF2["embarked"] == "Q")).sum() / (DF2["embarked"] == "Q").sum() * 100
```

```python
# 티켓 평균 비용 (물가)
S_price = DF2[(DF2["grade"] == "상") & (DF2["embarked"] == "S")]["per_fare"].mean()
C_price = DF2[(DF2["grade"] == "상") & (DF2["embarked"] == "C")]["per_fare"].mean()
Q_price = DF2[(DF2["grade"] == "상") & (DF2["embarked"] == "Q")]["per_fare"].mean()
```

```python
# 전체 티켓 비용의 평균 구하기
S_ticket_percentage = ((DF2["per_fare"] > DF2["per_fare"].mean()) & (DF2["embarked"] == "S")).sum() / (DF2["embarked"] == "S").sum() * 100
C_ticket_percentage = ((DF2["per_fare"] > DF2["per_fare"].mean()) & (DF2["embarked"] == "C")).sum() / (DF2["embarked"] == "C").sum() * 100
Q_ticket_percentage = ((DF2["per_fare"] > DF2["per_fare"].mean()) & (DF2["embarked"] == "Q")).sum() / (DF2["embarked"] == "Q").sum() * 100
```

```python
# 비교 대상으로 데이터프레임 생성하기
result = pd.DataFrame({"high_percentage" : [S_high_percentage, C_high_percentage, Q_high_percentage],
              "price_index" : [S_price, C_price, Q_price],
              "ticket_percentage" : [S_ticket_percentage, C_ticket_percentage, Q_ticket_percentage]}, 
              index = ["Southampton", "Cherbourg", "Queenstown"])
result
```

```python
# 점수 합산 함수 생성

def calculate_total_score(data):
    L = []
    for i in range(len(data)):
        L.append(0)

    for i in range(len(result.columns)):
        for j in range(len(result.index)):
            if (data[data.columns[i]][j] == data[data.columns[i]].max()):
                L[j] += 3
            elif (data[data.columns[i]][j] == data[data.columns[i]].median()):
                L[j] += 2
            elif (data[data.columns[i]][j] == data[data.columns[i]].min()):
                L[j] += 1
    return L
```

```python
# 결과 데이터프레임에 총 점수 추가

result["Total_score"] = calculate_total_score(result)
result
```

### 결론
- 탑승객 수 중에서 고위직 비율, 티켓의 평균 가격으로 추정한 물가, 티켓의 가격 평균보다 더 높은 가격을 쓴 사람의 비율로 점수를 매겼을 때 
- 프랑스의 Cherbourg의 점수가 가장 높으므로 3개의 나라 중에 프랑스의 Cherbourg가 그 당시 가장 부촌이라고 추정할 수 있다.
