### [데이터 전처리]
- Gender : 0, 1 => 남, 여
- GradeClass : 0.0, 1.0, 2.0, 3.0, 4.0 => A, B, C, D, F
- New_ParentalSupport 행 추가 = ParentalSupport + Tutoring (부모의 도움 정도가 같더라도, 과외를 시켜주는 부모가 좀 더 자녀에게 신경을 써줬다고 가정) => [상 중 하]로 분류
- Tutoring : 0, 1 => X, O

```python
data["Tutoring"] = data["Tutoring"].astype("int64")
```

```python
data["Gender"] = pd.cut(data["Gender"],
                        bins = 2,
                        labels = ["남", "여"])
data
```

```python
data["GradeClass"] = data["GradeClass"].astype("int64")
```

```python
data["GradeClass"] = pd.cut(data["GradeClass"],
                            bins = 5,
                            labels = ["A", "B", "C", "D", "F"])
data
```

```python
data["Ethnicity"].value_counts()
```

```python
data["Tutoring"] = pd.cut(data["Tutoring"],
               bins = 2,
               labels = ["X", "O"])
data
```

```python
data["new_ParentalSupport"] = pd.cut(data["new_ParentalSupport"],
                                      bins = 3,
                                      labels = ["하", "중", "상"])
data
```

```python
data2 = data[["ParentalEducation", "StudyTimeWeekly", "Absences", "Tutoring", "ParentalSupport", "GradeClass", "new_ParentalSupport"]]
data2
```

### [열 추가]
- Strictness 열 추가 : 출석율이 낮고, 공부량이 높을 수록 부모가 엄격하다는 가정


```python
data2["Strictness"] = (data2["Absences"] < data2["Absences"].mean()) & (data2["StudyTimeWeekly"] > data2["StudyTimeWeekly"].mean())
data2
```

```python
data2["Strictness"] = data2["Strictness"].astype("bool")
```

```python
data2["Strictness"] = pd.cut(data2["Strictness"],
                             bins = 2,
                             labels = ["X", "O"])
data2
```

#### [비교]
- 부모의 지원이 성적에 영향을 미치는가?
- 부모의 성격이 성적에 영향을 미치는가?
- 부모의 학력이 성적에 영향을 미치는가?

```python
# 눈으로 비교 불가 -> 비율로 따져서 비교하기

print(data2[data2["new_ParentalSupport"] == "상"]["GradeClass"].value_counts())
print(data2[data2["new_ParentalSupport"] == "중"]["GradeClass"].value_counts())
print(data2[data2["new_ParentalSupport"] == "하"]["GradeClass"].value_counts())
```

```python
def percentage_cal(data, column, condition, op):
    if op == "==":
        result = ((data[column] == condition) & (data["GradeClass"] <= "B")).sum() / (data[column] == condition).sum() * 100
    
    elif op == ">":
        result = ((data[column] > condition) & (data["GradeClass"] <= "B")).sum() / (data[column] > condition).sum() * 100    

    elif op == "<":
        result = ((data[column] < condition) & (data["GradeClass"] <= "B")).sum() / (data[column] < condition).sum() * 100    

    return result
```

```python
high_support_percentage = percentage_cal(data2, "new_ParentalSupport", "상", "==")
medium_support_percentage = percentage_cal(data2, "new_ParentalSupport", "중", "==")
low_support_percentage = percentage_cal(data2, "new_ParentalSupport", "하", "==")

print(F"지원을 많이 받은 성적 우수자 : {round(high_support_percentage, 2)}%")
print(F"지원을 적당히 받은 성적 우수자 : {round(medium_support_percentage, 2)}%")
print(F"지원을 조금 받은 성적 우수자 : {round(low_support_percentage, 2)}%")
```

```python
strict_parent_percentage = percentage_cal(data2, "Strictness", "O", "==")
non_strict_parent_percentage = percentage_cal(data2, "Strictness", "X", "==")

print(F"엄격한 부모님을 가진 성적 우수자 : {round(strict_parent_percentage, 2)}%")
print(F"온순한 부모님을 가진 성적 우수자 : {round(non_strict_parent_percentage, 2)}%")
```

```python
# 학사 이상, 이하의 부모
high_edu_parent_percentage = percentage_cal(data2, "ParentalEducation", 2, ">")
low_edu_parent_percentage = percentage_cal(data2, "ParentalEducation", 3, "<")

print(F"학사 학위 이상 부모님을 가진 성적 우수자 : {round(high_edu_parent_percentage, 2)}%")
print(F"학사 학위 미만 부모님을 가진 성적 우수자 : {round(low_edu_parent_percentage, 2)}%")
```
