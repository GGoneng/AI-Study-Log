```python
StudyTimeWeekly_1 = pd.cut(df['StudyTimeWeekly'],
                    bins = 3,
                    labels = ['하', '중', '상'])
```

```python
StudyTimeWeekly_1
```

```python
df['StudyTimeWeekly_1'] = StudyTimeWeekly_1
```

```python
df
```

```python
Absences_1 = pd.cut(df['Absences'],
                    bins = 3,
                    labels = ['하', '중', '상'])
```

```python
Absences_1
```

```python
df['Absences_1'] = Absences_1
```

```python
df
```

```python
GradeClass_1 = pd.cut(df['GradeClass'],
                    bins = 3,
                    labels = ['하', '중', '상'])
```

```python
GradeClass_1
```

```python
df['GradeClass_1'] = GradeClass_1
```

```python
df
```

```python
# StudyTimeWeekly 상,중,하에 따른 GradeClass의 평균
study_time_grade = df.groupby('StudyTimeWeekly_1')['GradeClass'].mean()
print(study_time_grade)
```

```python
# Absences 상,중,하에 따른 GradeClass의 평균
absences_correlation = df.groupby('Absences_1')['GradeClass'].mean()
print(absences_correlation)
```

```python
# Tutoring 여부에 따른 GradeClass의 평균
tutoring_grade = df.groupby('Tutoring')['GradeClass'].mean()
print(tutoring_grade)
```

```python
how = {    
    'GradeClass': np.mean,
}
df.groupby(['Absences_1', 'StudyTimeWeekly_1', 'Tutoring']).agg(how)
```
