- 결론: 학생의 학업 성취도에 영향을 미치는 요소에 대한 포괄적인 관점을 제공하므로 교육 연구, 예측 모델 개발, 통계 분석에 도움을 준다.

```python
# 학습 태도 : 주간공부시간 + 결석수
# 평귤 결석보다 낮고, 주간 평균 공부 시간보다 높은 성적 B 이상 비율
((df['Absences'] <= df['Absences'].mean()) & (df['StudyTimeWeekly'] >= df['StudyTimeWeekly'].mean()) & (df['GradeClass'] <= 1)).sum() / ((df['Absences'] <= df['Absences'].mean()) & (df['StudyTimeWeekly'] >= df['StudyTimeWeekly'].mean())).sum() * 100
```

```python
# 평귤 결석보다 높고, 주간 평균 공부 시간보다 낮은 학생이 성적 B 이상 받을 확률
((df['Absences'] >= df['Absences'].mean()) & (df['StudyTimeWeekly'] <= df['StudyTimeWeekly'].mean()) & (df['GradeClass'] <= 1)).sum() / ((df['Absences'] >= df['Absences'].mean()) & (df['StudyTimeWeekly'] <= df['StudyTimeWeekly'].mean())).sum() * 100
```
