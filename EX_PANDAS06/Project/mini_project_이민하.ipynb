```python
columns_name = data.columns.to_list()
L = []
for i in range(len(columns_name)):
    L.append(columns_name[i].strip())
data.columns = L

data.info()
```

```python
data.dropna(inplace = True)
data
```

```python
data.corr(numeric_only = True)
```

```python
corrSR = data.corr(numeric_only = True)["Alcohol"]
corrSR.drop(["Alcohol", "Year"], inplace = True)
corrSR
```

```python
columns = data.columns[3:].drop("Alcohol")
fig = plt.figure(figsize = (20, 20))
axes = fig.subplots(6, 3)

idx = 0

for i in range(len(axes)):
    for j in range(len(axes[0])):
        axes[i][j].scatter(data["Alcohol"], data[columns[idx]], label = round(corrSR.iloc[idx], 2), s = data["Year"] / 200, c = data["Year"],
             alpha = 0.5)
        axes[i][j].set_xlabel("Alcohol"); axes[i][j].set_ylabel(columns[idx])
        axes[i][j].legend()
        
        idx += 1
```

```python
# 벨라루스와 에스토니아는 술을 많이 먹는 국가이므로 이상치라고 보기 어려움
data[data["Alcohol"] > 17]
```

```python
fig = plt.figure(figsize = (15, 5))
ax1, ax2 = fig.subplots(1, 2, sharey = True)

high_gdp = data[data["GDP"] > data["GDP"].mean()]
low_gdp = data[data["GDP"] <= data["GDP"].mean()]

ax1.plot(high_gdp["Alcohol"], 'g-', label = round(high_gdp["Alcohol"].mean(), 2))
ax2.plot(low_gdp["Alcohol"], 'y-', label = round(low_gdp["Alcohol"].mean(), 2))

ax1.set_xticks(np.arange(len(high_gdp.index.unique())), high_gdp.index.unique(), rotation = 90)
ax2.set_xticks(np.arange(len(low_gdp.index.unique())), low_gdp.index.unique(), rotation = 90)

ax1.legend()
ax2.legend()

plt.suptitle("GDP별 알코올 섭취량 그래프", size = 15)
ax1.set_title("높은 GDP")
ax2.set_title("낮은 GDP")

plt.show()
```

```python
plt.figure(figsize = (5, 5))
plt.plot(high_gdp["Alcohol"], 'g-')
plt.plot(high_gdp["Life expectancy"] - 70 , 'r-')
plt.ylim()
plt.xticks(rotation = 90)
plt.title("높은 GDP를 가진 나라의 알코올 섭취량과 기대 수명의 관계")
plt.show()
```

```python
plt.figure(figsize = (7, 7))
plt.plot(low_gdp["Alcohol"], 'g-')
plt.plot(low_gdp["Life expectancy"] - 70, 'r-')
plt.xticks(rotation = 90)
plt.title("낮은 GDP를 가진 나라의 알코올 섭취량과 기대 수명의 관계")
plt.show()
```

```python
high_alcohol = data[data["Alcohol"] > data["Alcohol"].mean()]
low_alcohol = data[data["Alcohol"] <= data["Alcohol"].mean()]
```

```python
plt.figure(figsize = (5, 5))
plt.pie([high_alcohol["percentage expenditure"].sum(), low_alcohol["percentage expenditure"].sum()], labels = ["high_alcohol", "low_alcohol"],
         autopct = '%.1f%%', wedgeprops = {"linewidth":2, 'edgecolor':"white"}, counterclock = False, startangle = 90, explode = [0.01, 0])
plt.title("[ 알코올 별 병원 비용 ]")
plt.legend()
plt.show()
```

```python
high_alcohol_developing_ratio = round(((high_alcohol["Status"] == "Developing").sum() / high_alcohol["Status"].count()), 2) * 100
high_alcohol_developed_ratio = round(((high_alcohol["Status"] == "Developed").sum() / high_alcohol["Status"].count()), 2) * 100
low_alcohol_developing_ratio = round(((low_alcohol["Status"] == "Developing").sum() / low_alcohol["Status"].count()), 2) * 100
low_alcohol_developed_ratio = round(((low_alcohol["Status"] == "Developed").sum() / low_alcohol["Status"].count()), 2) * 100
```

```python
plt.pie([high_alcohol_developing_ratio, low_alcohol_developing_ratio, high_alcohol_developed_ratio, low_alcohol_developed_ratio], labels = ["high_alcohol_developing", "low_alcohol_developing", "high_alcohol_developed", "low_alcohol_developed"],
         autopct = '%.1f%%', wedgeprops = {"linewidth":2, 'edgecolor':"white"})
plt.legend()
plt.tight_layout()
plt.title("나라 발전 별 알코올 섭취 비교")
plt.show()
```
