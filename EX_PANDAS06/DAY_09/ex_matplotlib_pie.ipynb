#### [ 파이 그래프 ]
- 컬럼(변수)의 데이터의 분포 비율을 시각화

```python
# [1] 모듈 로딩
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# [2] 데이터 준비
F_DATA = r"C:\Users\KDP-2\OneDrive\바탕 화면\Python\EX_PANDAS06\Data\auto_mpg.csv"
```

```python
# [3] CSV => DataFrame으로 저장
dataDF = pd.read_csv(F_DATA)
dataDF
```

```python
# - 컬럼별 타입 확인
dataDF.dtypes
```

```python
# [4] 특정 컬럼의 데이터 비율
# - 자동차 생산 국가에 대한 비율
dataDF['origin'].astype("category")
```

```python
# - 자동차 생산 국가에 대한 비율
round(dataDF['origin'].value_counts() / dataDF['origin'].count(), 2)
```

```python
originSR = dataDF["origin"].value_counts()
x = originSR.index.to_list()
y = originSR.values.tolist()
x, y
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - bar 그래프
origin_bar = plt.bar(x, y, color = ["red", "green", "blue"], edgecolor = "yellow")
# x축 눈금 설정
plt.xticks([1, 2, 3], ["US", "EU", "JP"], rotation = 45, fontsize = 15)

# 막대에 라벨을 설정
plt.bar_label(origin_bar, y, label_type = 'center')
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(y, labels = ['US', 'EU', 'JP'], autopct = '%.1f%%')
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(y, labels = ['US', 'EU', 'JP'], autopct = '%.1f%%', counterclock = False, 
        startangle = 90)
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(y, labels = ['US', 'EU', 'JP'], autopct = '%.1f%%', counterclock = False, 
        startangle = 90, explode = [0, 0, 0.1])
plt.show()
```

```python
# - 자동차 생산 국가에 대한 비율 시각화 - pie 그래프
plt.pie(y, labels = ['US', 'EU', 'JP'], radius = 1, autopct = '%.1f%%', counterclock = False, 
        startangle = 90, wedgeprops = {"linewidth":1, 'width':0.9, 'edgecolor':"white"})
plt.pie(y, labels = ['US', 'EU', 'JP'], radius = 0.7, autopct = '%.1f%%', counterclock = False, 
        startangle = 90, wedgeprops = {"linewidth":1, 'width':0.9, 'edgecolor':"white"})
plt.show()
```
