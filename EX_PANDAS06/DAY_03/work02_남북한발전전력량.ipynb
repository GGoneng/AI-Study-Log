```python
import pandas as pd
import openpyxl
```

```python
def row_sum(data):

    total = []
    sum = 0

    for i in range(len(data)):
        for j in range(2, len(data.iloc[i])):
            sum += data.iloc[i, j]
        total.append(sum)
        sum = 0

    data["합계"] = total
```
