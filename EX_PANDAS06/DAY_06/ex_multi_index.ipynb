```python
# [1] 모듈 로딩
import pandas as pd
import df_util as util
```

```python
# [2] 데이터 준비
file1 = "test_data.csv"
file2 = "test_data2.csv"
```

```python
# [3] CSV ==> DataFrame으로 저장

data1 = pd.read_csv(file1)
data2 = pd.read_csv(file2)
```

```python
# [4] 데이터 확인 => 기본정보
util.checkDataFrame(data1, "data1")
util.checkDataFrame(data2, "data2")
```

```python
# [5] 행 인덱스 설정
# [5-1] data1의 첫번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])

df11 = data1.set_index(["index"])
df11.index.name = None
df11.index

# "A"행 데이터를 추출
df11.loc["A"]
```

```python
# 'A', 'C'행 추출
df11.loc[['A', 'C']]
```

```python
# [5-2] data1의 첫번째, 두번째 컬럼을 행 인덱스로 설정 ==> set_index([컬럼1, 컬럼2])

df12 = data1.set_index(["index", "col1"])
print(df12.index)

# ("A", 8) 행 추출
df12.loc[('A', 8)]
```

```python
df12.loc[[('A', 8), ('A', 1)]]
```

```python
# 멀티 인덱스 속성들 확인
mIdx = df12.index
print(F"mIdx : {mIdx.nlevels}")
print(F"mIdx : {mIdx.levels}")
print(F"mIdx : {mIdx.names}")
print(F"mIdx : {mIdx.levshape}")

# mIdx.names = [None, None]
# print(F"mIdx : {mIdx.names}")

# mIdx.names[0] = "A"ArithmeticError
mIdx.set_names("a", level = 0)
```

```python
# 멀티 인덱스에서 level0이 동일한 행 추출
# - "A"
df12.loc["A"]
```

```python
# 멀티 인덱스에서 level1이 동일한 행 추출
# - 8
# df12.loc[8]

df12.xs(8, level = 1)
```

```python
# 멀티 인덱스에서 행 추가
# - 인덱스에 행 추가 => 변수명.loc["행 인덱스"] = 값

df12.loc[("D", 9), :] = [9, 9]

sr1 = pd.Series({"col2":10, "col3":10})
sr1.index = ["D", 10]
sr1.T
```
