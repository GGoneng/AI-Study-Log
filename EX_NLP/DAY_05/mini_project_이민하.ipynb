```python
from urllib.request import urlopen
from urllib.request import Request

from bs4 import BeautifulSoup

import pandas as pd
import re
```

```python
def crawling(classifier):
    L = []
    for idx in range(1, 335):
        url = f'https://www.health.kr/researchInfo/disease_detail.asp?idx={idx}'
        urlrequest = Request(url, headers = {'User-Agent' : 'Mozilla/5.0'})

        html = urlopen(urlrequest)
        bs = BeautifulSoup(html, 'html.parser')
        table = bs.find_all('p')
        table2 = bs.find_all('td') 

        if classifier in table2[2].text:
            for row in table[1:-3]:
                L.append(row.text)

        elif classifier in table2[3].text:
            for row in table[1:-3]:
                L.append(row.text)
                
        else:
            pass
    return L
```

```python
text_data = crawling('순환')
```

```python
L2 = crawling('호흡')
```

```python
text_data.extend(L2)
```

```python
text_data
```

```python
text = ''.join(text_data)
text
```

```python
txt_list = []
text_list = text.split('.')

for text in text_list:
    regex_pattern = re.compile('[^ㄱ-ㅎ가-힣 ]+')
    pre_text = re.sub(regex_pattern, ' ', text)
    txt_list.append(pre_text)

txt_list
```

```python
txt_list = pd.DataFrame(txt_list)
txt_list
```

```python
txt_list.to_csv('text_data.csv', index = False, encoding = 'utf-8', header = None)
```
