```python
import pickle
import re
import pandas as pd
import numpy as np
import os
```

```python
with open('text1.pkl', 'rb') as f:
    text = pickle.load(f)

with open('title1.pkl', 'rb') as f:
    title = pickle.load(f)

with open('opinion1.pkl', 'rb') as f:
    opinion = pickle.load(f)

with open('director1.pkl', 'rb') as f:
    director = pickle.load(f)
    
    
```

```python
# 피클 파일이 있는 폴더 경로
folder_path = './data'
title = []
text = []
opinion = []
director = []

# 'list'가 올바르게 파이썬의 기본 타입인지 확인
print(type(list))  # 기본적으로는 <class 'list'>가 출력되어야 합니다.

# 폴더 경로가 존재하는지 확인
if not os.path.exists(folder_path):
    print(f"폴더 경로가 존재하지 않습니다: {folder_path}")
else:
    # 폴더 내 모든 파일 탐색
    for filename in os.listdir(folder_path):
        
            # 'title'이 포함된 파일 처리
            if 'title' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
                file_path = os.path.join(folder_path, filename)  # 파일 경로 생성
                with open(file_path, 'rb') as f:
                    data = pickle.load(f)
                
                 
                    title.extend(data)


            # 'text'가 포함된 파일 처리
            elif 'text' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
                file_path = os.path.join(folder_path, filename)
                with open(file_path, 'rb') as f:
                    data = pickle.load(f)
                  
                    text.extend(data)


            # 'opinion'이 포함된 파일 처리
            elif 'opinion' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
                file_path = os.path.join(folder_path, filename)
                with open(file_path, 'rb') as f:
                    data = pickle.load(f)
                  
                    opinion.extend(data)
  

            # 'director'가 포함된 파일 처리
            elif 'director' in filename and (filename.endswith('.pickle') or filename.endswith('.pkl')):
                file_path = os.path.join(folder_path, filename)
                with open(file_path, 'rb') as f:
                    data = pickle.load(f)
                   
                    director.extend(data)
 





print("title:", len(title))
print("text:", len(text))
print("opinion:", len(opinion))
print("director:", len(director))
```

```python
opinion_1 = []

for txt in opinion:
    result = re.sub(r'\s+', ' ', txt)

    L = []
    
        
    for t in result.split('|'):
        L.append(t.strip())

    opinion_1.append(L)

```

```python
i = 0

for t in text:
    text[i] = t.strip()
    i += 1
```

```python
title_2d = []
for tit in title:
    title_2d.append([tit])

director_2d = []
for dir in director:
    director_2d.append([dir])

text_2d = []
for t in text:
    text_2d.append([t])
```

```python
for i in range(len(title)):
    title_2d[i].extend(opinion_1[i])
    title_2d[i].extend(director_2d[i])
    title_2d[i].extend(text_2d[i])
```

```python
column_name = ['제목', '유형', '구분', '장르', '상영 시간', '등급', '국적', '감독', '시놉시스', '?']
df = pd.DataFrame(title_2d, columns = column_name)
```

```python
df = df[df['?'].isnull()]

df.drop('?', axis = 1, inplace = True)
df.dropna(inplace= True)
```
